```python
import torch
from torch import tensor
```

```python
def matmul(a,b):
    ar,ac = a.shape # n_rows * n_cols
    br,bc = b.shape
    assert ac==br
    c = torch.zeros(ar, bc)
    for i in range(ar):
        for j in range(bc):
            for k in range(ac): c[i,j] += a[i,k] * b[k,j]
    return c
```

```python
m1 = torch.randn(5,28*28)
m2 = torch.randn(784,10)
```

```python
%time t1=matmul(m1, m2)
```

```python
%timeit -n 20 t2=m1@m2
```
